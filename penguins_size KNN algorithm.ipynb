```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(penguins.drop('sex_MALE',axis = 1))
```

```python
scaled_features = scaler.transform(penguins.drop('sex_MALE',axis = 1))
```

```python
penguins_size = pd.DataFrame(scaled_features,columns = penguins.columns[:-1])
penguins_size
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(scaled_features, penguins['sex_MALE'], test_size = 0.2, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred[:10]
```

```python
# Evaluate and Predictions 
from sklearn.metrics import confusion_matrix , classification_report
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

```python
# Choose the optimal K by Elbow Method 
# (1,40) 40 iterations do randomly taking 1 to 40 

error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,40),error_rate, color = 'blue', linestyle = 'dashed', marker = 'o', markerfacecolor = 'red', markersize=10)
plt.title("Error Rate Vs K Value")
plt.xlabel("K Values")
plt.ylabel("Error Rate")
plt.show()
```

```python
#Now with K = 24 because see in above graph like like in 24 above the variations of lines not comming down so based on that take k number
# the error should be very less near to 0 
knn = KNeighborsClassifier(n_neighbors = 24)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
pred[:10]
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

```python
#Now with K = 25

knn = KNeighborsClassifier(n_neighbors = 25)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
pred[:10]
```

```python
print(classification_report(y_test, pred))
```

#### The overall accuracy of the model is 0.88, meaning that 88% of the predictions were correct.
